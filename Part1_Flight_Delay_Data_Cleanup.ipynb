```python
import pandas as pd
from pathlib import Path

# Supress Warnings
import warnings
warnings.simplefilter(action = "ignore")
```

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
data = Path("Resources/Jan_2020_ontime.csv")
data_df = pd.read_csv(data)

# Review the DataFrame
data_df.head()
```

```python
# Check the dimensionality of the DataFrame.
data_df.shape
```

## Data Cleanup

```python
# Drop columns which won't be used
data_df = data_df.drop([ 
                    'OP_CARRIER_AIRLINE_ID', 
                    'OP_CARRIER', 'TAIL_NUM', 
                    'OP_CARRIER_FL_NUM',
                    'ORIGIN_AIRPORT_ID',
                    'ORIGIN_AIRPORT_SEQ_ID', 
                    'DEST_AIRPORT_ID', 
                    'DEST_AIRPORT_SEQ_ID', 
                    'DEP_TIME',
                    'CANCELLED','ARR_TIME',
                    'DIVERTED',
                    'Unnamed: 21'
                   ], axis = 1)

data_df.head()
```

```python
# Check the dimensionality of the DataFrame.
data_df.shape
```

```python
# Generate descriptive statistics of DataFrame
data_df.describe()
```

```python
# Check the columns' data types
data_df.dtypes
```

```python
# Number of NaN values in df for each column
data_df.isna().sum()
```

```python
# dropped NaN values in DataFrame
data_df = data_df.dropna()
data_df.shape
```

```python
# verifying if NaN values were dropped
data_df.isna().sum()
```

```python
# Number of Unique elements in DataFrame
data_df.nunique()
```
